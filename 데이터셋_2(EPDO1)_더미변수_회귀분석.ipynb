# 1.x를 도로종류랑 피해자신체상해주부위 합쳐서 회귀 시행

```python
x=df[['피해자신체상해주부위','도로종류','행동유형가해자','차량용도가해자_중분류']]
x = pd.get_dummies(x,drop_first=True)
ols = OLS(y,x).fit().summary()
ols
```
