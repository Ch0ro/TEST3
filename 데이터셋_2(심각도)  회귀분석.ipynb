## 2.전체

```python
x=df.drop(['심각도'],axis=1)
x = pd.get_dummies(x,drop_first=True)
x= x.drop(['연령가해자_기타/불명','법규위반가해자_기타','피해자신체상해주부위_기타','피해자신체상해주부위_불명','차량용도가해자_중분류_기타불명',
          '차량용도가해자_중분류_불명','도로종류_기타','도로형태_기타','도로형태_기타/불명','도로형태_불명'],axis=1)
ols = OLS(y,x).fit().summary()
ols
```

#### P값 유의하지 않은 것 삭제

```python
x=x.drop(['사고유형_길가장자리구역통행중','법규위반가해자_교차로 통행방법 위반','법규위반가해자_보행자 보호의무 위반','법규위반가해자_안전운전 의무 불이행'
          ,'행동유형가해자_주ㆍ정차중','행동유형가해자_주행 중 대기','피해자신체상해주부위_얼굴','피해자신체상해주부위_하퇴부'
         ,'차량용도가해자_중분류_특수여객(장의)','도로형태_단일로','도로형태_주차장','발생시각_23-01시'],axis=1)
```

```python
ols = OLS(y,x).fit().summary()
ols
```

#### R로 steopAIC 시행해보기 위해 파일 출력
